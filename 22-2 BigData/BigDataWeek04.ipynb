```python
# DataFrame - Using Dictionary
import pandas as pd
data = {
    'age' : [29, 33, 39],
    'height' : [169, 170, 183],
    'weight' : [65, 74, 81]
}
indexName = pd.Series(['ari', 'insung', 'thanin'])
frame2 = pd.DataFrame(data, index  = indexName)
frame2
```

```python
# Search from DataFrame
# row
print(frame2.loc['ari'])
print(frame2.iloc[2])
# column
print(frame2['age'])
print(frame2.age)
```

```python
# add a new column
frame2_col_added = pd.DataFrame(frame2, columns = ['age', 'height', 'weight', 'now_col'])
frame2_col_added
```

```python
# add a new row
frame2_row_added = frame2_col_added.copy()
frame2_row_added.loc['inguk'] = [37, 180, 68, 'new1']
frame2_row_added
```

```python
# save DataFrame as csv format
frame2_row_added.to_csv('df_to_csv.csv')
```

```python
# read csv as DataFrame
df_from_csv = pd.read_csv('df_to_csv.csv')
df_from_csv
```

```python
# set index name
df_from_csv.set_index('Unnamed: 0', inplace = True)
df_from_csv
```
