```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('california_house.csv')
data
```

```python
data.info()
```

Q1. Drawing Multiple Subplots

```python

plt.figure(figsize = (12, 8))


plt.subplot(2,3,1)
plt.hist(data['HouseAge'])
plt.title('HouseAge_hist')

plt.subplot(2,3,2)
plt.hist(data['Latitude'])
plt.title('Latitude_hist')

plt.subplot(2,3,3)
plt.hist(data['Longitude'])
plt.title('Longitude_hist')



plt.subplot(2,3,4)
plt.boxplot(data['HouseAge'], vert=False)
plt.title('HouseAge_box')

plt.subplot(2,3,5)
plt.boxplot(data['Latitude'], vert=False)
plt.title('Latitude_box')

plt.subplot(2,3,6)
plt.boxplot(data['Longitude'], vert=False)
plt.title('Longitude_box')

plt.plot()   


plt.show()
```

Q2. Calculating Skewness and Kurtosis

```python
fig, axs = plt.subplots(figsize = (16, 12), ncols = 3, nrows = 3)

x_features = ['MedInc','HouseAge','AveRooms','AveBedrms','Population', 'AveOccup', 'Latitude', 'Longitude', 'MedHouseVal']
for i, feature in enumerate(x_features):
    
    skew = data[feature].skew()
    kurt = data[feature].kurtosis()
    print('{0} - Skewness: {1}, Kurtosis: {2}'.format(feature, skew, kurt))
    
    
    logsales = np.log1p(data[feature])
    
    row = int(i/3)
    col = i % 3
    
    sns.histplot(logsales, kde=True, bins=30, ax = axs[row][col])
```

Q3. Which column is the most asymmetric? Which column is the sharpest?

A3.

The skewness and kurtosis of the 'AveBedrms' are the largest

(Skewness: 4.365356897236684, Kurtosis: 35.63490710212813)

So it's the most asymmetrical and sharpest column
