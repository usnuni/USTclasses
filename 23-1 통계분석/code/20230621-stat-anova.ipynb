```python
import numpy as np
from scipy.stats import f
import matplotlib.pyplot as plt
```

```python
sample_means = [713, 756.93, 698.07, 562.02]
sample_vars  = [46.55**2, 40.34**2, 37.20**2, 39.87**2]
```

## STEP 01: Set up Hypotheses

- $H_0$ : All of the popuation means are equal
- $H_a$ : At least one population has different mean from the others. 

## STEP 02: Calculate Test Statistics

- $F=(MSTr)/(MSE)$
- MSTr = ?
- MSE  = ? 

- Numerator = MSTr ~ variance of sample means

```python
I = 4
J = 6
```

```python
MSTr = np.var(sample_means, ddof=1) * J
print(MSTr)
```

- Denominator = MSE = Mean of individual sample variances

```python
MSE = np.mean(sample_vars)
print(MSE)
```

```python
F = MSTr / MSE
print(F)
```

## STEP 03. Calculate Rejection region with given significance level

- significance level

```python
alpha = 0.05
```

- We need to define F-distribution for test statistic. 

```python
nu1 = I - 1
nu2 = I*(J-1)
print(nu1, nu2)
```

```python
rv = f(dfn=nu1, dfd=nu2)
```

```python
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots(1, 1, figsize=(15,5))
ax.plot(x, rv.pdf(x), color="blue")
plt.show()
```

- Now we can calculate the rejection region for upper-sided test.

```python
boundary = rv.ppf(0.95)
print(boundary)
```

- Rejection region : $(3.098, \infty)$

## STEP 04. Compare test statistic and rejection region

- 25.09 is inside the rejection region. 
- So we reject the null hypothesis; At least one population has different population mean.
