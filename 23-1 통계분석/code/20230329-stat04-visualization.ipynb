```python
data['Age'].fillna(data['Age'].mean(), inplace=True)
```

```python
data_fare = data['Fare'].values  # Save column data as NumPy Array
```

```python
data_age = data['Age'].values 
```

- Let us visualize the 'Fare' column!

# Visualization : Matplotlib Library

- https://matplotlib.org/

```python
import matplotlib.pyplot as plt
```

- You start to define how many subplots are included in your Figure.
- By using subplots, you can define your figure and your subplots. 
- fig = variable for your overall figure
- ax (axes) = variable for each subplot
- figsize = (width, height)

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 3))
ax.plot(data_fare)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
ax[0,0].plot(data_age)
ax[1,1].plot(data_fare)
plt.show()
```

```python
ax.shape
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
ax[0].plot(data_age)
ax[1].plot(data_fare)
plt.show()
```

## Histogram

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
ax[0].hist(data_age,  bins=20, edgecolor="white", color="green", alpha=0.5)
ax[1].hist(data_fare, bins=20, edgecolor="white")
ax[0].grid()
ax[1].grid()
plt.show()
```

## Boxplot

```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 5))
ax[0].boxplot(data_age, vert=False, patch_artist=True)
ax[1].boxplot(data_fare, vert=False, patch_artist=True)
plt.show()
```
