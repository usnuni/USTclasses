## 0. String

- Sequential (Ordered) Collection of Characters (alphabets in English, special characters, blank space, alphabets in other language, etc.)
- Each element of string has unique position.
- Example : hello vs ohell - 1 h, 1 e, 2 l, and 1 o. 
- Each element of string has index (integer) number, which defines its position (location). 
- The first index of your string (list) starts from 0.

```python
a = "Good Morning!"
```

```python
len(a)
```

- indexing : accessing to a single element by using the corresponding index number

```python
a[2]
```

```python
a[5]
```

- slicing : taking some part of your string out of your original string
- Q: What if I want to have "Good" out of "Good Morning!"?
- For slicing, you need to specify two indices (positions): Where your slice starts and where your slice ends.
- Syntax : (variable name)[(1st index):(2nd index)] : The element corresponding to the 2nd index is NOT included in the slice. Your slice stops right before  2nd index. 

```python
a[0:3]
```

```python
a[0:4]
```

### Q. Why do we need to use NumPy array?

- NumPy array is more efficient and faster than List when you do some computation. 
- When you have very small size of data, you cannot recognize difference in computational speeds between NumPy Array and List.
- But, if you need to handle a sort of **big data**, NumPy array boosts up calculations.

### Variance Calculation

- If your data is population, then you use the population variance. 
- If your data is sampled from a population, you need to use the sample variance. 
- Difference between population and sample variances? : prefactor $\frac{1}{N}$ vs $\frac{1}{n-1}$

#### CASE 1: Your data is population data. 

```python
np.var(a)
```

```python
np.std(a)
```

```python
np.sqrt(np.var(a))
```

#### CASE 2: Your data is sampled from the population

```python
np.var(a, ddof=1)
```

```python
np.std(a, ddof=1)
```
