# Interval Estimation

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

# Case 01 : Population follows normal distribution with known population variance

- We want to make interval estimation for population mean. 

## STEP00. Sampling and Calculating Sample Mean

```python
mu, sigma = 10, 0.200
n = 20
```

```python
population = norm(loc=mu, scale=sigma)
```

```python
sample = population.rvs(size=n)
```

```python
print(sample)
```

```python
sample_mean = np.mean(sample)
print(sample_mean)
```

## STEP01. Set-up Confidence Level

- confidence level = $1 - \alpha = 0.95$, $\alpha=0.05$ 

```python
alpha = 0.05
```

```python
rv = norm(loc=0, scale=1.0)
```

## STEP02. Calculate Upper and Lower Bounds of Confidence Interval

- You can use (1) z-critical value or (2) percentiles of the standard normal distribution.

```python
lower_bound, upper_bound = rv.ppf(alpha*0.5), rv.ppf(1-alpha*0.5)
print(lower_bound, upper_bound)
```

- confidence interval for $Z=\frac{\bar{X}-\mu}{\sigma/\sqrt{n}}$ : -1.96 < Z < 1.96

## STEP03. Obtain the intervarl for the population mean $\mu$

- $Z=\frac{\bar{X}-\mu}{\sigma/\sqrt{n}}$ $\leftrightarrow$ $\mu = \bar{X} - Z\frac{\sigma}{\sqrt{n}} $.

```python
sample_mean + lower_bound*sigma/np.sqrt(n), sample_mean + upper_bound*sigma/np.sqrt(n)
```

-----

```python
fig, ax = plt.subplots(1,1, figsize=(5,20))

ax.plot([mu,mu], [0.0, 1.0], color="black", linestyle=":")

for i in range(100):
    
    # STEP00 : Sampling
    sample = population.rvs(size=20)
    sample_mean = np.mean(sample) # \bar{x}
    
    # STEP01 : confidence interval for Z with 1-alpha level
    alpha = 0.05
    rv = norm(loc=0, scale=1.0)
    lower_bound, upper_bound = rv.ppf(alpha*0.5), rv.ppf(1-alpha*0.5)
    
    # STEP02 : confidence interval for population mean
    lower = sample_mean + lower_bound*sigma/np.sqrt(n)
    upper = sample_mean + upper_bound*sigma/np.sqrt(n)
       
    ax.plot([lower, upper], [0.01*i, 0.01*i], color="blue", linestyle="-")    

interval_size = upper - lower
print(interval_size)

plt.show()
```
